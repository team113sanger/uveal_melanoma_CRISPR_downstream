sgRNA-level Area Under ROC and Area Under Precision-Recall Curves across cell lines

```r
library(tidyverse)

sgRNA_level_tot <- read_tsv('../results/sgRNA_level_metrics.tsv', col_types = cols())

ggplot(sgRNA_level_tot, aes(x = cell_line, y = AUC, fill = metric)) +
  geom_bar(stat='identity', position = 'dodge', color = NA) +
  theme_bw() +
  labs(x = 'Cell line', y = 'AUROC/AUPrR', fill = 'Metric', title = 'sgRNA-level metrics') +
    theme(plot.title = element_text(hjust = 0.5, size = 16), panel.border = element_blank(), 
    panel.grid.major = element_blank(), panel.grid.minor = element_blank(), 
    aspect.ratio = 1,
    axis.line = element_line(colour = "black"),
    axis.text = element_text(size = 12, colour = "black"),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.title = element_text(size = 14),
    legend.text = element_text(size = 12),
    legend.title = element_text(size = 14))
```
